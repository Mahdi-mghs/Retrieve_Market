```python
model_ser = dirty_data['model_name']
real_model = []
for ind,mod in enumerate(model_ser):
    temp_str = mod.split('|')[-1]

    real_model.append(temp_str.split(' ا ')[-1])

print(len(real_model))
# print(real_model)
```

```python
for ind,i in enumerate(real_model):
    real_model[ind] = real_model[ind].replace('\u200c', '').replace('\u200d', '').replace('لپ تاپ', '').replace('لپتاپ', '')
# print(real_model)
```

```python
# Create a dictionary to map Persian brand names to English brand names
brand_mapping = {
    'ایسوس': 'Asus',
    'لنوو': 'Lenovo',
    'اپل': 'Apple',
    'اچ پی': 'HP',
    'ام اس آی': 'MSI',
    'ایسر': 'Acer',
    'مایکروسافت': 'Microsoft',
    'مک بوک': 'MacBook'
}


# Function to replace Persian brand names with English brand names
def replace_persian_brands(laptop_name):
    for persian_brand, english_brand in brand_mapping.items():
        laptop_name = laptop_name.replace(persian_brand, english_brand)
    return laptop_name

# Apply the brand name replacement to the list
english_laptop_list = [replace_persian_brands(name) for name in real_model]

# Print the updated list
for name in english_laptop_list:
    print(name)
```

```python
real_model = pd.Series(english_laptop_list)
for ind, i in enumerate(real_model):
    if 'استوک' in i:
        real_model[ind] = np.nan
        continue
    elif 'ا' in i:
        real_model[ind] = np.nan
        continue
    elif 'ت' in i:
        real_model[ind] = np.nan
        continue
    real_model[ind] = i.replace('مدل', '').replace('Laptop', '').replace('  ', ' ')
    real_model[ind] = i.strip()

real_model.isnull().sum()
        
```
